```python
import pandas as pd
a=pd.read_csv("HRDatasets.csv")
print(a)
```

```python
print(a.isnull().sum())
```

```python
#Change Format
a['Salary']=a['Salary'].astype(float)
a['DateofHire']=pd.to_datetime(a['DateofHire'],format='%d-%m-%Y')
a['DOB']=pd.to_datetime(a['DOB'],format='%d-%m-%Y')
a['DateofTermination']=pd.to_datetime(a['DateofTermination'],format='%d-%m-%Y')
a['LastPerformanceReview_Date']=pd.to_datetime(a['LastPerformanceReview_Date'],format='%d-%m-%Y')
print(a.describe())
```

```python
#counting values of categorical column
print("Marital Data")
print(a['MaritalDesc'].value_counts())

print("\n Department Data")
print(a['Department'].value_counts())

print("\n Position Data")
print(a['Position'].value_counts())

print("\n Manager data")
print(a['ManagerName'].value_counts())
```

```python
#calculating total Employees
total_employees=len(a)
print(f"Total Employees:{total_employees}")
```

```python
#Filling Missing Values
a['DateofTermination']=a['DateofTermination'].fillna(value="Currently Working") #Fill with currently working
print(a['DateofTermination'])
```

```python
#Calculating Old emplotees(sum of currently working employees)
#count no of emp who have left 
employees_left_count=(a['DateofTermination']!='Currently Working').sum() #uisng not equal to 104 left job
print(f"Number of Employees who have left:{employees_left_count}")
```

```python
#currently working employees 
employees_working_count=(a['DateofTermination']=='Currently Working').sum() #uisng not equal to 104 left job
print(f"Number of Employees currently working:{employees_working_count}")
```

```python
#calculate total department
total_departments=a['Department'].nunique() #nunique(Not allow Duplicate Values)
print(f'Total Departments:{total_departments}')
```

```python
#Calculate Avg Salary
average_salary=a['Salary'].mean()
print(f'Average Salary:{average_salary}')
```

```python
#Avg age of Employee
import datetime as dt
a['Age']=(dt.datetime.now()-a['DOB']).dt.days/365.25 #only coln of dob we create new coln age and minus from current days(.25 adjust leap yr)
average_age=a['Age'].mean()
print(f'Average Age:{average_age:.2f}')
```

```python
#Avg years with comapny
from datetime import datetime
a['YearsWithCompany']=(dt.datetime.now()-a['DateofHire']).dt.days/365.25 
average_years_with_company=a['YearsWithCompany'].mean()
print(f'Average years:{average_years_with_company:.2f}')
```

```python
#Attrition rate(count of leaving people)
attrition_rate=(employees_left_count/total_employees)*100 
print(f'Attrition Rate:{attrition_rate:.2f}%')
```

```python
#Making graph of total employees per department
import matplotlib.pyplot as plt
#Group by department and count the no.of employees in each department
department_counts=a['Department'].value_counts()
plt.figure(figsize=(10,6))
department_counts.plot(kind='bar',color='skyblue')
plt.title("Total No.of Employees in Each Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
#Distribution of salary
plt.figure(figsize=(12,6))
sns.histplot(a['Salary'],bins=30,kde=True)
plt.title("Distribution of Salary")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```
